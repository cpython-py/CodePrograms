# ndarray
## 1.ndarray特性 # NOTE
### 多维性

```python
import numpy as np
```

```python
def br():
    print('-' * 60)
def printr(content):
    print(content, '-'*60, sep='\n', end='\n')
```

```python
arr = np.array(5) # 创建一个0维ndarray数组
print('内容:', arr)
type(arr)
print('arr的维度:', arr.ndim) # 打印arr的维度
```

```python
arr = np.array([1, 2, 3]) # 创建一个1维ndarray数组
print('内容:', arr)
type(arr)
print('arr的维度:', arr.ndim) # 打印arr的维度
```

```python
arr = np.array([ [1, 2, 3], [4, 5, 6] ]) # 创建一个2维ndarray数组
print('内容:', arr)
type(arr)
print('arr的维度:', arr.ndim) # 打印arr的维度
```

### 同质性 # NOTE

```python
arr = np.array([1, 'hello'])
print(arr)
```

```python
arr = np.array([1, 2.5])
print(arr)
```

## 3.ndarray的创建

```python
# 基础的创建方法
lst1 = [1, 2, 3]
arr = np.array(lst1, dtype=np.float64)
print(arr)
```

```python
# copy
arr1 = np.copy(arr)
print(arr1)
arr1[0] = 8
print(arr1)
print(arr)
```

```python
# 预定义形状
# 全0 全1 未初始化 固定值
# 全0
arr = np.zeros((2, 3))
print(arr)
br()
arr = np.zeros((2, 3), dtype=int)
print(arr)
br()
arr = np.zeros((200, ))
print(arr)
br()
# 全1
arr = np.ones((5, 8))
print(arr)
br()
# 未初始化
arr = np.empty((4, 3))
printr(arr)
# 固定值
arr = np.full((3, 5), 2025)
printr(arr)
```

```python
# _like模仿形参数组的形状
arr1 = np.zeros_like(arr)
arr2 = np.empty_like(arr)
arr3 = np.ones_like(arr)
arr4 = np.full_like(arr, 2026)
printr(arr1)
printr(arr2)
printr(arr3)
printr(arr4)
```

```python
# 等差数列
arr = np.arange(1, 10, 1)
printr(arr)
# 等间隔数列
arr = np.linspace(1, 10, 5)
printr(arr)
```

```python
# 练习
arr = np.arange(0, 101, 25)
arr2 = np.linspace(0, 100, 5, dtype=int)
print(arr)
print(arr2)
```

```python
# 对数间隔数列
arr = np.logspace(0, 4, 3, base=2, dtype=int) # base: default = 10
print(arr)
```

```python
# 特殊矩阵
# 单位矩阵: 主对角线上的数字为1, 其他的数组为0
arr = np.eye(3, 4, dtype=int)
print(arr)
br()
arr = np.eye(3, dtype=int)
print(arr)
# 对角矩阵: 主对角线上的数字非0, 其余的为0
arr = np.diag([1, 2, 3])
br()
print(arr)
```

```python
# 随机数组的生成
# 生成0到1之间的随机浮点数(均匀分布)
arr = np.random.rand(2, 3)
print(arr)
br()
# 生成指定范围的随机浮点数
arr = np.random.uniform(3, 6, (2, 3))
print(arr)
br()
# 生成指定范围的随机整数
arr = np.random.randint(3, 30, (3, 6))
print(arr)
br()
# 生成随机数列(正太分布: 两边小,中间大) 
arr = np.random.randn(2, 3) # 范围: (-3, 3)
print(arr)
```

```python
# 设置随机种子: 保证随机生成的数值相同
np.random.seed(20)
arr = np.random.randint(1, 10, (2, 5))
print(arr)
```

4.ndarray的数据类型 # NOTE

布尔: bool

整形: int uint

浮点型: float

复数: complex

```python
from numpy import uint


arr = np.array([1, 0, 2, 0], dtype='bool') # dtype = 'bool' | bool | np.bool
print(arr)
arr = np.array([1, 0, 255, 0], dtype=uint)
print(arr)
```
