## 1.ndarray特性 # NOTE

```python
def br():
    print('-' * 60)
def printr(content):
    print(content, '-'*60, sep='\n', end='\n')
```

## 3.ndarray的创建

```python
# 基础的创建方法
lst1 = [1, 2, 3]
arr = np.array(lst1, dtype=np.float64)
print(arr)
```

```python
# copy
arr1 = np.copy(arr)
print(arr1)
arr1[0] = 8
print(arr1)
print(arr)
```

```python
# 预定义形状
# 全0 全1 未初始化 固定值
# 全0
arr = np.zeros((2, 3))
print(arr)
br()
arr = np.zeros((2, 3), dtype=int)
print(arr)
br()
arr = np.zeros((200, ))
print(arr)
br()
# 全1
arr = np.ones((5, 8))
print(arr)
br()
# 未初始化
arr = np.empty((4, 3))
printr(arr)
# 固定值
arr = np.full((3, 5), 2025)
printr(arr)
```

```python
# _like模仿形参数组的形状
arr1 = np.zeros_like(arr)
arr2 = np.empty_like(arr)
arr3 = np.ones_like(arr)
arr4 = np.full_like(arr, 2026)
printr(arr1)
printr(arr2)
printr(arr3)
printr(arr4)
```

```python
# 等差数列
arr = np.arange(1, 10, 1)
printr(arr)
# 等间隔数列
arr = np.linspace(1, 10, 5)
printr(arr)
```

```python
# 练习
arr = np.arange(0, 101, 25)
arr2 = np.linspace(0, 100, 5, dtype=int)
print(arr)
print(arr2)
```

```python
# 对数间隔数列
arr = np.logspace(0, 4, 3, base=2, dtype=int) # base: default = 10
print(arr)
```

```python
# 特殊矩阵
# 单位矩阵: 主对角线上的数字为1, 其他的数组为0
arr = np.eye(3, 4, dtype=int)
print(arr)
br()
arr = np.eye(3, dtype=int)
print(arr)
# 对角矩阵: 主对角线上的数字非0, 其余的为0
arr = np.diag([1, 2, 3])
br()
print(arr)
```

```python
# 随机数组的生成
# 生成0到1之间的随机浮点数(均匀分布)
arr = np.random.rand(2, 3)
print(arr)
br()
# 生成指定范围的随机浮点数
arr = np.random.uniform(3, 6, (2, 3))
print(arr)
br()
# 生成指定范围的随机整数
arr = np.random.randint(3, 30, (3, 6))
print(arr)
br()
# 生成随机数列(正太分布: 两边小,中间大) 
arr = np.random.randn(2, 3) # 范围: (-3, 3)
print(arr)
```

```python
# 设置随机种子: 保证随机生成的数值相同
np.random.seed(20)
arr = np.random.randint(1, 10, (2, 5))
print(arr)
```

## 5.索引与切片 # NOTE 

```python
# 一维数组
arr = np.random.randint(1, 50, 50)
print(arr)
```

```python
print(arr[0])
print(arr[:])
print(arr[2:5])
print(arr[slice(2, 11, 2)])
print(arr[arr>40])
print(arr[ (arr>30) & (arr<40) ])
print(arr[ (arr>40) | (arr<10) ])
```

```python
# 二维数组
arr = np.random.randint(1, 100, (4, 8))
print(arr)
```

```python
print(arr[2, 3])
print(arr[:,:])
print(arr[1,:])
print(arr[1,2:5])
print(arr[arr>50])
print(arr[2][arr[2]>50])
print(arr[:,3])
```

## 6.ndarray的运算 #NOTE

```python
# 算术运算
a = np.array([1, 2, 3])
b = np.array([2, 3, 4])
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)
c = [1, 2, 3]
d = [4, 5, 6]
print(c + d)
```

```python
# 二维数组
a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([[4, 5, 6], [1, 2, 3]])
print(a + b)
print(a - b)
print(a * b)
print(a / b)
print(a ** 2)
print(a + 3)
```

```python
# 广播机制
a = np.array([1, 2, 3])
b = np.array([[1], [2], [3]])
print(a)
print(b)
print(a + b)
print(a - b)
```

```python
# 矩阵运算
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 4], [6, 8]])
print(a @ b)
```

## numpy中的常用函数 # NOTE

## 2.统计函数
求和, 平均值, 中位数, 标准差, 方差 \
最大值, 最小值 \
分位数, 累计和, 累计积

```python
arr = np.random.randint(1, 20, 8)
print(arr)
```

```python
# 求和
print(np.sum([1, 2, 3]))
# 平均值
print(np.mean([1, 2, 3]))
# 中位数
print(np.median([1, 2, 4]))
print(np.median([1, 2, 3, 4]))
print(np.median([4, 1, 2]))
```

```python
# 计算方差
print(np.var([1, 2, 3]))
# 计算标准差
print(np.std([1, 2, 3]))
```

```python
# 计算最大值, 最小值
print(np.max([1, 2, 100, 2, 199, 32]))
print(np.min([1, 2, 100, 2, 199, 32]))
arr = np.array([1, 2, 3, 4])
print(np.percentile(arr, 20))
```

```python
arr = np.array([1, 2, 3])
# 累计和
print(np.cumsum(arr)) # 0+1, 1+2, 3+3
print(np.prod(arr)) # 0*1, 1*2, 2*3
```

## 3.比较函数
大于, 小于, 等于 \
与, 或, 非 \
是否有一个True, 是否都为True, 自定义条件

```python
print(np.greater([3, 4, 5, 6, 7], 4)) # 是否大于
print(np.less([3, 4, 5, 6, 7], 4)) # 是否小于
print(np.equal([3, 4, 5, 6, 7], 4)) # 是否等于
print(np.equal([1, 2, 3], [4, 2, 3])) # !形状相同
print(np.logical_and([1, 1], [1, 0])) # 逻辑与
print(np.logical_or([0, 0], [1, 1])) # 逻辑或
print(np.logical_not([1, 0]))  # 逻辑非
print(np.any([1, 0, 0, 1])) # 至少一个为真
print(np.all([1, 0, 0, 1])) # 全部为True
arr = np.array([1, 2, 3, 4, 5])
print(np.where(arr>3, arr, 0)) # ?自定义 (条件, 符合, 不符合)
score = np.random.randint(20, 100, 50)
print(np.select([score>80, (score>=60) & (score<=80), score<60], ('优秀', '及格', '不及格'), default='未知')) #?选择 (条件, 返回的结果)

```

## 4.排序函数

```python
np.random.seed(0)
arr = np.random.randint(1, 100, 20)
print(arr)
arr.sort() # ?返回None, 从小到大
print(arr)
print(np.sort(arr)) # ?返回一个新数组, 不影响原始数组
print(np.argsort(arr)) # ?返回排序的索引
```

```python
print(np.unique([1, 3, 1, 2])) # 去重, 并且排序
print(np.concatenate(([1, 2, 3], [4, 5, 6]))) # 拼接数组
print(np.split(arr, 4)) # 分成4份, 不可等分报错
print(np.split(arr, [6, 12, 18])) # 切割的位置
print(np.reshape(arr, [4, 5])) # 重新设置形状 不等分报错
```

```python
# [28, 30, 29, 31, 32, 30, 29]
arr = np.array([28, 30, 29, 31, 32, 30, 29]) # 七天的温度
print('%.2f' % np.mean(arr)) # 平均温度
print(np.max(arr)) # 最高温度
print(np.min(arr)) # 最低温度
print(len(arr[arr>30])) # 温度大于30的天数
```

```python
# [85, 90, 78, 92, 88]
scores = np.array([85, 90, 78, 92, 88]) # 成绩
print(np.mean(scores)) # 平均成绩
print(np.median(scores)) # 中位数
print(np.std(scores)) # 标准差
print(scores / 10) # 转换满分为10分
```

```python
# [[1, 2], [3, 4]], [[5, 6], [7, 8]]
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(A + B)
print(A * B)
print(A @ B)
```

```python
np.random.seed(20)
integer = np.random.randint(0, 10, (3, 4))
print(integer)
print(np.max(integer, axis=0))
print(np.min(integer, axis=1))
print(np.where(integer%2==1, -1, integer))
```

```python
arr = np.arange(1, 13)
print(arr)
arr = np.reshape(arr, [3, 4])
print(np.sum(arr, axis=1))
print(np.mean(arr, axis=0))
print(np.reshape(arr, (12)))
```

```python
np.random.seed(10)
arr = np.random.randint(0, 20, (5, 5))
print(arr)
print(arr[arr>10])
print(np.where(arr>10, 0, arr))
```

```python
# [120, 135, 110, 125, 130, 140]
money = np.array([120, 135, 110, 125, 130, 140])
print(np.sum(money))
print(np.mean(money))
print(np.var(money))
print(np.argmax(money) + 1)
print(np.argmin(money) + 1)
```
