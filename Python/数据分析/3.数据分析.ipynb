```python
# 缺失值的处理 
import numpy as np
s = pd.Series([1, 2, np.nan, None, pd.NA, 4, 5])
print(s)
# 查看是否是缺失值
print(s.isna())
print(s.isnull())
print(s.isna().sum())  # 查看缺失值的个数
print(s.dropna())  # 删除缺失值
# 填充缺失值
df = pd.read_csv('data/weather_withna.csv')
print(df.tail())
print(df.isna().sum(axis=0))
print(df.fillna({'temp_max':2, 'wind':2.5}))
print(df.ffill().tail())  # ffill() 使用前面的相邻数据填充
print(df.bfill().tail())  # bfill() 使用后面的相邻数据填充
```

```python
import pandas as pd
import numpy as np
data = {
    'name': ['alice', 'alice', 'bob', 'alice', 'jack', 'bob'],
    'age': [26, 25, 30, 25, 35, 30],
    'city': ['NY', 'NY', 'LA', 'NY', 'SF', 'LA']
}
df = pd.DataFrame(data)
print(df)
print(df.duplicated())  # 一整条记录都是一样的标记为重复, 返回bool
print(df.drop_duplicates(subset=['name'], keep='last')) # 根据name去除重复, 指定列, 保持最后出现的
```

```python
# 数据类型的转换
df = pd.read_csv('data/sleep.csv')
print(df)
print(df.dtypes)
df['age'] = df['age'].astype('int16')
print(df.tail())
print(df.dtypes)
print(df.gender)
df['gender'] = df['gender'].astype('category')
print(df.dtypes)
print(df.gender)
df['is_male'] = df['gender'].map({'Female': True, 'Male': False})  # map()映射
print(df.is_male)
```

```python
# 数据分箱
import pandas as pd
df = pd.read_csv('data/employees.csv')
df1 = df.head(10)[['employee_id','salary']]
# print(df1)
print(pd.cut(df1['salary'], bins=2))  # bins分成2个箱(区间)
print(pd.cut(df1['salary'], bins=3).value_counts())
print(pd.cut(df1['salary'], bins=[0, 10000, 20000, 30000]).value_counts())  # bins自定义箱(区间)
print(pd.cut(df1['salary'], bins=[0, 10000, 20000, 30000], labels=['低', '中', '高']))  # bins自定义箱(区间), labels指定箱的标签
# qcut等频分箱
print(pd.qcut(df1['salary'], q=3))
```

```python
# 时间数据处理
import pandas as pd
d = pd.Timestamp('2026-04-12 12:24') # 时间戳
print(d)
print(type(d))
print(d.year, d.month, d.day, d.hour, d.minute, d.second)
print(d.quarter) # 季度
print(d.is_month_end) # 是否是月底
print(d.day_name()) # 星期几
print(d.to_period('D')) # 转换为天级时间戳
print(d.to_period('Q')) # 转换为季度时间戳
print(d.to_period('M')) # 转换为月级时间戳
print(d.to_period('Y')) # 转换为年级时间戳
print(d.to_period('W')) # 转换为周时间戳
```
