```python
import pandas as pd
```

```python
# 创建Series数组
s = pd.Series([1, 2, 3, 4, 5])
print(s)
# 自定义索引
s = pd.Series([10, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'])
print(s)
# 定义name
s = pd.Series([10, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'], name='月份')
print(s)
# 通过字典来创建
s = pd.Series({'a':1, 'b':2, 'c':3, 'd':4, 'e':5})
print(s)
# 继承
s1 = pd.Series(s, index=['a', 'c'])
print(s1)
```

```python
# Series的属性
s = pd.Series([10, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'], name='月份')
print(s.index)
print(s.values)
print(s.ndim, s.shape, s.size, s.dtype, s.name)
print(s.loc['A']) # 显示索引 按标签名 更推荐
print(s.loc['A':'C'])
print(s.iloc[0]) # 隐式索引 按位置 更推荐
print(s.iloc[0:2])
print(s.at['B']) # 显示索引 不支持切片
print(s.iat[4]) # 隐式索引 不支持切片
```

```python
# 访问数据
s = pd.Series([1, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'], name='月份')
s['F'] = 6
print(s)
print(s['A'])
print(s[s<3])
print(s.head()) # * 默认获取表格中前5行的信息 可以在括号内写入数字, 如写入3, 会获取前3行的数据
print(s.tail()) # * 默认获取表格中后5行的信息 可以在括号内写入数字, 如写入2, 会获取后2行的数据
```

```python
# Series常见函数
import numpy as np
s = pd.Series([10, 2, np.nan, None, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E', 'F', 'G'])
print(s)
print(s.head())
print(s.describe()) # * 获取所有描述性信息 count 会忽略缺失值NaN
print(s.count())
print(s.keys()) # * 获取索引对象
print(s.index) # * 和keys()效果相同
print(s.isna()) # * 判断某个元素是否为缺失值
print(s.isin([4])) # * 判断某个元素是否存在 需要传入列表
print(s.mean(), s.std(), s.var(), s.min(), s.max(), s.sum())
print(s.median())
print(s.sort_values()) # * 按值排序
print(s.sort_index()) # * 按索引排序
print(s.quantile(0.25)) # * 传入的数字为 [0, 1]
s['H'] = 2
print(s.mode()) # * 众数
print(s.value_counts()) # * 值出现的次数 忽略缺失值
print(s.drop_duplicates(), type(s.drop_duplicates())) # * 去重 缺失值会变成1个
print(s.unique(), type(s.unique())) # * 去重 返回的数据类型和drop_duplicates()不一样, 返回numpy的ndarray类型
print(s.nunique()) # * 返回去重后的个数
```

```python
# 案例.学生成绩统计
np.random.seed(42)
scores = pd.Series(np.random.randint(50, 101, 10), index=['Student'+str(i) for i in range(1, 11)])
print(scores)
print(scores.mean()) # 平均分
print(scores.max()) # 最高分
print(scores.min()) # 最低分
print(scores[scores>scores.mean()].size) # 高于平均分的人数
```

```python
# 案例.温度数据分析
temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33], index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
print(temperatures[temperatures>30].size) # 温度大于30的天数
print(temperatures.mean()) # 平均温度
print(temperatures.sort_values(ascending=False)) # 按温度从高到低排序 # ? ascending是否升序, 不影响原始数组
print(*temperatures.diff().abs().sort_values(ascending=False).keys()[:2].tolist()) # 获取温度变化最大的两天 # ? 1=>diff() 获取相邻两个元素的后-前的差值 2=>tolist()转换成列表 3=> * 去掉列表的中括号
```

```python
# 案例.股票价格分析
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2], index=pd.date_range('2026-04-10', periods=10))
# ? pandas.date_range() 日期序列, periods 序列长度
print(prices)
print(prices.pct_change()) # 计算每日收益率, # ?收益率 = (今天的/昨天的) -1
print(prices.pct_change().idxmax(), prices.pct_change().idxmin(), sep='\t|\t') # 收益率最高和最低的日期 # * 1.idx- 获取对应的索引
print(prices.std()) # 波动率
```

```python
# 案例.销售数据分析
sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220], index=pd.date_range('2026-04-10', periods=12, freq='MS'))
print(sales)
print(sales.resample('QS').mean()) # 季度的平均销量 # * resample() 重新采样 'QS' 按每个季度
print(sales.resample('QS').mean().idxmax()) # 平均销量最高的季度
print(sales.pct_change()) # 月环比的增长率
Gt0 = sales.pct_change() > 0
print(Gt0[Gt0.rolling(3).sum()==3].keys().tolist()) # 找出连续增长超过两个月的月份
```

```python
# 案例.每小时销售数据分析
np.random.seed(42)
hours_sales = pd.Series(np.random.randint(0, 100, 24), index=pd.date_range('2026-4-10', periods=24, freq='h'))
day_sales = hours_sales.resample('D').sum()
print(day_sales) # 计算每日总销售额
business_hours_sales = hours_sales[(hours_sales.index.hour>=8) & (hours_sales.index.hour<=22)].sum()
print('-----' * 10)
print(business_hours_sales / (day_sales - business_hours_sales)) # 计算营业时间和非营业时间的销售额比例
print(*hours_sales.nlargest(3).keys().tolist(), sep='\n') # 找出销售额最高的三个小时
```

## DataFrame对象

```python
# 创建DataFrame对象
# 通过Series对象创建
np.random.seed(100)
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series(['a', 'b', 'c', 'd', 'e'])
df = pd.DataFrame({'A': s1, 'B': s2})
print(df, type(df))
print(df['A'], type(df['A']))
# 通过字典来创建
df = pd.DataFrame(
    {
    'student_id': [1, 2, 3, 4, 5],
    'name': ['张三', '李四', '王五', '赵六', '钱七'],
    'age': [18, 19, 20, 21, 22],
    'score': [90.3, 80, 30.1, 78.3, 94]
    },
    index=np.random.randint(1, 100000, size=5),
    columns=['student_id', 'name', 'score', 'age']) # * columns设置列的顺序
print(df)
```

```python
# DataFrame的属性和方法 和Series的属性和方法 基本一致
print(df.index, type(df.index))
print(df.values, type(df.values))
print(df.dtypes)
print(df.shape)
print(df.ndim)
print(df.size)
print(df.columns) # * 列标签
print(df.T)
print(df.loc[77656, 'student_id'])
print(df.loc[77656, :])
print(df.loc[:, 'name'])
print(df.iloc[:, 0])
print(df.at[65616, 'age'])
print(df.iat[0, 0])
print(df[['name', 'age']])
print(df.head(3))
print(df.tail(3))
print(df.score>70) # * 可以通过自己设置的属性来筛选数据
print(df.student_id)
print(df[df.score>60])
print(df[(df.score>60) & (df.age>18)])
print(df.sample()) # * 随机抽样, 默认抽样1行, 可以通过n参数来指定抽样行数
print(df.sample(n=3)) # * 随机抽样, 抽样3行
```

```python
# 案例.学生成绩分析
data = {
    '姓名': ['张三', '李四', '王五', '赵六', '钱七'],
    '数学': [80, 91, 75, 85, 95],
    '英语': [90, 85, 95, 88, 92],
    '物理': [85, 90, 88, 92, 95],
}

scores = pd.DataFrame(data, index=[1, 2, 3, 4, 5])
print(scores)
scores['总分'] = scores[['数学', '英语', '物理']].sum(axis=1)
scores['平均分'] = scores[['数学', '英语', '物理']].mean(axis=1)
print(scores)
print('---' * 10)
print(scores[(scores['数学']>95) | (scores['英语']>90)])
scoresSortByTotalScore = scores.sort_values('总分', ascending=False)
print(scoresSortByTotalScore.head(3))
```

```python
# 案例.销售数据分析
data = {
    '产品名称': ['A', 'B', 'C', 'D', 'E'],
    '单价': [100, 200, 300, 400, 500],
    '销售量': [10, 20, 30, 40, 50],
}
df = pd.DataFrame(data)
print(df['单价'] * df['销售量'])
print((df['单价'] * df['销售量']).max())
df['销售额'] = df['单价'] * df['销售量']
print(df.sort_values('销售额', ascending=False))
print(df.describe())
```
