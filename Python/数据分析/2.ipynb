```python
import pandas as pd
# TODO Series构造函数:pd.Series()
GDP = [80855,77388,68024,47251,40471]
city = ['GD','JS','SD','ZJ','HN']
info = pd.Series(GDP,index=city)
print(info)
print(type(info))
```

```python
# TODO 创建一个一维数组当做Series的值
import numpy as np
GDP = np.array([80855,77388,68024,47251,40471])
rank = [1,2,3,4,5]
info = pd.Series(GDP,index=rank)
print(info)
print(type(info))
```

```python
# TODO 传入常量
s = pd.Series(6,index=['a','b','c','d'])
print(s)
```

```python
# TODO 索引访问和索引标签访问
info = pd.Series([80855,77388,68024,67251,40471],
index=['GD','JS','SD','ZJ','HN'])
print(info[0])
print(info['JS'])
```

```python
# TODO Series的三种常用属性:
    # 1.dtype 2.values 3.index
print(info.dtype)
print(info.values)
print(info.index)
```
