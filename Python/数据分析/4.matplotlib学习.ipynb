```python
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体
rcParams['font.sans-serif'] = ['SimHei']

# 创建图标, 绘制大小
plt.figure(figsize=(10, 5))  # 画布
# 前6行固定要写 和plt.show()
# 添加标题
plt.title('2026年销售趋势', color='aqua', fontsize=20)
month = ['1月', '2月', '3月', '4月']
sales = [100, 150, 80, 130]
# 添加坐标轴的标签
plt.xlabel('月份', fontsize=20, color='green')
plt.ylabel('销售额(万元)', fontsize=20, color='green')
plt.plot(month, sales, color='orange', linewidth=2, linestyle='--', marker='o', label='产品A')
# marker 标记点的样式 label 图例
# 添加图例
plt.legend(loc='upper left')
# 添加网格线
# plt.grid(True)
plt.grid(axis='y', alpha=0.2, color='blue', linestyle='--')  # * alpha 透明度 linestyle 网格线的样式 -- 虚线
# 设置刻度字体大小
plt.xticks(rotation=45, fontsize=12)  # rotation 旋转角度
plt.yticks(rotation=45, fontsize=12)  
# 设置y轴范围
plt.ylim(0, 160)
# 在每个数据点上显示数值
for x, y in zip(month, sales):
    plt.text(x, y+3, str(y), ha='center', va='bottom', fontsize=10)  # ha 水平对齐 va 垂直对齐 s 显示数值 x x轴显示的坐标, y y轴显示的坐标
# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体
rcParams['font.sans-serif'] = ['SimHei']

# 创建图标, 绘制大小
plt.figure(figsize=(10, 5))  # 画布
subjects = ['语文', '数学', '英语', '科学']
scores = [85, 92, 78, 88]
# 绘制柱状图
plt.bar(subjects, scores, label='小红', color='gold', width=0.5)
plt.title('小红2026年成绩分布柱状图', color='red', fontsize=16)
plt.xlabel('科目', color='gray')
plt.ylabel('成绩', color='gray')
plt.legend(loc='upper right')
plt.grid(axis='y', alpha=0.1, color='gray', linestyle='--')
plt.ylim(0, 100)
for x, y in zip(subjects, scores):
    plt.text(x, y+1, str(y), ha='center', va='bottom', fontsize=10)
plt.tight_layout()  # 自动调整子图参数，使子图之间的间距更小
plt.show()
```

```python
# 绘制饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体
rcParams['font.sans-serif'] = ['SimHei']

# 创建图标, 绘制大小
plt.figure(figsize=(10, 5))  # 画布
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
plt.pie(times, labels=things, autopct='%1.1f%%', startangle=90, colors=colors)  
# autopct: 自动添加百分比标签 colors: 配色方案 wedgeprops: 配置环形图的属性 
plt.title('一天的时间分布', color='green', fontsize=30)
plt.text(0, 0, '总计: \n100%', ha='center', va='bottom')
plt.tight_layout() 
plt.show()
```

```python
# 绘制环形图
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体
rcParams['font.sans-serif'] = ['SimHei']

# 创建图标, 绘制大小
plt.figure(figsize=(10, 5))  # 画布
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
plt.pie(times, labels=things, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops={'width': 0.6})  
# autopct: 自动添加百分比标签 colors: 配色方案 wedgeprops: 配置环形图的属性 
plt.title('一天的时间分布', color='green', fontsize=30)
plt.text(0, 0, '总计: \n100%', ha='center', va='bottom')
plt.tight_layout() 
plt.show()
```

```python
# 绘制爆炸式饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams  # 字体
rcParams['font.sans-serif'] = ['SimHei']

# 创建图标, 绘制大小
plt.figure(figsize=(10, 5))  # 画布
things = ['学习', '娱乐', '运动', '睡觉', '其他']
times = [6, 4, 1, 8, 5]
explode = [0.2, 0, 0, 0, 0.2]  # 设置突出块的位置
colors = ['#66b3ff', '#99ff99', '#ffcc99', '#ff9999', '#ff4499']
plt.pie(times, labels=things, autopct='%1.1f%%', startangle=0, colors=colors, explode=explode, shadow=True)
# explode: 设置突出块的位置 shadow: 添加阴影
plt.title('一天的时间分布', color='green', fontsize=30)
plt.tight_layout() 
plt.show() 
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import random
rcParams['font.sans-serif'] = ['SimHei']
plt.figure(figsize=(10, 8)) 
x = []
y = []
for i in range(1000):
    tmp =  random.uniform(0, 10)
    x.append(tmp)
    y.append(2*tmp + random.gauss(0, 2))
plt.title('x-y的关系图', color='aqua', fontsize=20)
plt.xlabel('X', fontsize=20, color='green')
plt.ylabel('Y', fontsize=20, color='green')
plt.scatter(x, y, color='blue',alpha=0.5, s=10, label='数据点')  # s 点的大小
plt.legend(loc='upper left')
plt.grid(axis='y', alpha=0.2, color='blue', linestyle='--')  # * alpha 透明度 linestyle 网格线的样式 -- 虚线
plt.xticks(rotation=45, fontsize=12)
plt.yticks(rotation=45, fontsize=12)  
plt.ylim(0, 30)
plt.plot([0, 10], [0, 20], color='#ff5700', label='拟合线')


plt.show()
```

```python
# 箱线图
import matplotlib.pyplot as plt

# 模拟 3 门课的成绩
data = {
    '语文': [82, 85, 88, 70, 90, 76, 84, 83, 95],
    '数学': [75, 80, 79, 93, 88, 82, 87, 89, 92],
    '英语': [70, 72, 68, 65, 78, 80, 85, 90, 95]
}
plt.figure(figsize=(8, 6))
plt.boxplot(data.values(), tick_labels=data.keys())

plt.title("各科成绩分布（箱线图）")
plt.ylabel("分数")
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# 多个图的绘制方法
import matplotlib.pyplot as plt
# 要绘图的数据
month = ['1','2','3','4']
sales = [100,150,80,130]

# 动态图表的生成
# f1 = plt.subplot(2,2,1) #生成一个子图 行 列 索引
f1 = plt.subplot(221) #生成一个子图 行 列 索引
f1.plot(month,sales)
f2 = plt.subplot(2,2,2)
f2.bar(month,sales)
f3 = plt.subplot(2,2,3)
f3.scatter(month,sales)
f4 = plt.subplot(224)
f4.barh(month,sales)
```

```python
# 1. 导入库
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = ['SimHei']

# 2. 导入数据
df = pd.read_csv('data/weather.csv')
df.head()
# 绘制气温的趋势变化图
df['date'] = pd.to_datetime(df['date'])
df=df[df['date'].dt.year==2015]
plt.figure(figsize=(15,10))
plt.plot(df['date'],df['temp_max'],label='最高气温')
plt.plot(df['date'],df['temp_min'],label='最低气温')
plt.title('2015年气温趋势变化图')
plt.xlabel('日期')
plt.ylabel('气温')
plt.legend()

# 绘制降水量的直方图
df['temp_mean'] = (df['temp_max'] + df['temp_min'])/2
plt.figure(figsize=(15,10))
plt.plot(df['date'],df['temp_mean'],label='平均气温')
plt.title('2015年气温趋势变化图')
plt.xlabel('日期')
plt.ylabel('气温')
plt.legend()
# 绘制降水量的直方图
plt.hist(df['precipitation'],bins=5)
```
