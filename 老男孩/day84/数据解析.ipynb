```python
import requests
```

```python
fp = open('sanguo.text', 'w', encoding='utf-8')
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/145.0.0.0 Safari/537.36'}
main_url = 'https://www.shicimingjv.com/wenxue/sanguoyanyi/'
page_text = requests.get(main_url, headers=headers).text
soup = BeautifulSoup(page_text, 'lxml')
a_list = soup.select('.sec-panel-body > ul > li > a')
for a in a_list:
    title = a.string
    detail_url = a['href']

    detail_page_text = requests.get(detail_url, headers=headers).text
    detail_soup = BeautifulSoup(detail_page_text, 'lxml')
    content = detail_soup.select(".entry-content.clearfix")[0].text
    fp.write(title+'\t'+content+'\n')
    print(title, '下载成功!!!')
    
    
```
