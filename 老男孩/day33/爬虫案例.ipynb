HTTP状态码:
    200 : ok
    400 : bad request
    404 : not found

```python
import requests
# requests.get()方法,后面的参数为网页的url地址,返回一个response响应对象
url = requests.get("http://www.pinduoduo.com")
print(url)

# response.status_code属性,返回网页的http状态码
print(url.status_code)
# response.apparent_encoding属性,返回网页的编码格式
res = url.apparent_encoding
print(res)
# 设置编码集防止乱码
url.encoding = res
# response.text属性,返回网页的内容
res = url.text
print(res)
```
