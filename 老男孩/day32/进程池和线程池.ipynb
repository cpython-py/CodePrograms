```python
from concurrent.futures import ProcessPoolExecutor

# 进程池 ProcessPoolExecutor
def func(i):
    print(i)
    print("start..进程号{}".format(os.getpid()))
    time.sleep(3)
    print("end..进程号{}".format(os.getpid()))
    return i

if __name__ == "__main__":
    # 创建进程池对象
    """默认参数是os.get.cpu_count(),即CPU的核心数"""
    p = ProcessPoolExecutor()
    
    # 异步提交任务
    """submit(方法, 参数1, 参数2, ...)"""
    for i in range(10):
        obj = p.submit(func, i)
        print(obj)
        
```

```python
from concurrent.futures import ProcessPoolExecutor
import os, time

def func(i):
    print(f"Task {i} started in process {os.getpid()}")
    time.sleep(3)
    print(f"Task {i} finished in process {os.getpid()}")
    return f"Result of task {i}"

if __name__ == "__main__":
    p = ProcessPoolExecutor()
    
    futures = []
    for i in range(10):
        future = p.submit(func, i)
        futures.append(future)
        print(future)

    # 等待所有任务完成并获取结果
    for future in futures:
        try:
            result = future.result()  # 阻塞直到任务完成
            print(result)
        except Exception as e:
            print(f"Task failed with exception: {e}")
```
