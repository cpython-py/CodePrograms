# 线程队列
导入方式:import queue
线程常用队列:queue, LifoQueue, DriorityQueue.
queue: 先进先出
LifoQueue: 后进先出
DriorityQueue: 按照优先级排序
方法和进程队列类似
##
put 存放,超过列表长度会堵塞
get 获取,超过列表长度会堵塞
put_nowait 存放,超过列表长度会报错
get_nowait 获取,超过列表长度会报错

```python
# Queue
from queue import Queue
q = Queue()
q.put(100)
print(q.get())
# print(q.get()) 报错,队列为空

q = Queue(3)
q.put(200)
q.put(300)
q.put(400)
# q.put_nowait(500) 报错,队列已满


```

```python
# LifoQueue 先进后出, 后进先出 (栈)
from queue import LifoQueue
lq = LifoQueue()
lq.put(110)
lq.put(120)
lq.put(119)
print(lq.get())
print(lq.get())
print(lq.get())
```

```python
# PriorityQueue 优先级队列
"""按照优先级顺序进行排序存放(默认从小到大)"""
from queue import PriorityQueue
pq = PriorityQueue()
```

```python
# 对数字,按照从小到大排序
pq.put(100)
pq.put(19)
pq.put(-90)
pq.put(88)

print(pq.get())
print(pq.get())
print(pq.get())
print(pq.get())
```

```python
# 对字母,按照ASCII编码排序
pq.put("wangwen")
pq.put("sunjian")
pq.put("wangwei")

print(pq.get())
print(pq.get())
print(pq.get())
```

```python
# 对容器,将每一个容器的第一个元素进行比较,再对第二个元素进行比较,以此类推
pq.put((22, "wangwen"))
pq.put((67, "wangyuhan"))
pq.put((3, "zhangfengyong"))
pq.put((3, "liuyubo"))

print(pq.get())
print(pq.get())
print(pq.get())
print(pq.get())   
```
