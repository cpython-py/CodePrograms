```python
import socket as sk
sk = sk.socket()
sk.bind(('localhost', 8080))
sk.listen()
while True:
    conn, addr = sk.accept()
    res = conn.recv(1024).decode()
    path = res.split(" ")[1]
    
    if path == '/':
        with open('ceshi.html', 'rb') as fp:
            data = fp.read()
    elif path == '/1.jpg':
        with open('1.jpg', 'rb') as fp:
            data = fp.read()
    elif path == '/xx.css':
        with open('xx.css', 'rb') as fp:
            data = fp.read()
    elif path == '/xx.js':
        with open('xx.js', 'rb') as fp:
            data = fp.read()

    print()
    conn.send(b'HTTP/1.1 200 ok\r\n\r\n')
    conn.send(data)
    
    conn.close()
```
